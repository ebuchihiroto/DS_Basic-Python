## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df0 = pd.read_csv(r"C:\Users\Hiroto\DS_Basic-Python\data\hist_data.csv")
plt.figure(figsize=(4, 2.5))
plt.hist(df0["aデータ"], bins = 30,  color="orange", alpha = 0.5, label = ["aデータ"])
plt.hist(df0["bデータ"], bins = 30, color="lightgreen", alpha = 0.5, label = ["bデータ"])
plt.title("aデータとbデータの温度に関する度数分布", fontsize = 8)
plt.xlabel("温度(℃)", fontsize = 8)
plt.ylabel("度数", fontsize = 8)
plt.legend(fontsize = 8)
plt.show()

```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df1 = pd.read_csv(r"C:\Users\Hiroto\DS_Basic-Python\data\scatter_data.csv",)
class0 = df1[df1["class"] == 0]
class1 = df1[df1["class"] == 1]
class2 = df1[df1["class"] == 2]
class0_x, class0_y = class0["length (cm)"], class0["width (cm)"]
class1_x, class1_y = class1["length (cm)"], class1["width (cm)"]
class2_x, class2_y = class2["length (cm)"], class2["width (cm)"]
plt.scatter(class0_x, class0_y, color = "red", label = "class0")
plt.scatter(class1_x, class1_y, color = "blue", label = "class1")
plt.scatter(class2_x, class2_y,color = "green", label = "class2")
plt.xlabel("length (cm)")
plt.ylabel("width (cm)")
plt.title("クラス０～２のlenghtとwidthに関する散布図")
plt.legend(loc = "upper right")
plt.grid()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df2 = pd.read_csv(r"C:\Users\Hiroto\DS_Basic-Python\data\line_data.csv")
fig, axes = plt.subplots(1, 2, figsize = (12, 4))
plt.subplots_adjust(wspace = 0.3)
axes[0].plot(df2["期間"], df2["件数"], color = "blue", linestyle = "-", marker = "o")
axes[0].set_title("期間ごとの件数の推移")
axes[0].set_xlabel("期間")
axes[0].set_ylabel("件数")
axes[0].grid()
axes[0].tick_params(axis = 'x', rotation = 45)
axes[1].bar(df2["期間"], df2["費用"], color = "blue", label = "費用")
axes[1].set_title("期間ごとの費用の推移")
axes[1].set_xlabel("期間")
axes[1].set_ylabel("費用(円)")
axes[1].tick_params(axis = 'x', rotation = 45)
axes[1].grid()
plt.show()  
```
